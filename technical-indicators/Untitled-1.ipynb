```python
import yfinance as yf

# Define the tickers for US and UK indices
us_indices = ['^GSPC', '^DJI', '^IXIC']
uk_indices = ['^FTSE']

# Download historical data for US indices
us_data = yf.download(us_indices, start='2020-01-01', end='2024-08-15')
us_data=us_data.sort_values(by='Date', ascending=False)
# Download historical data for UK indices
uk_data = yf.download(uk_indices, start='2020-01-01', end='2024-08-15')
uk_data=uk_data.sort_values(by='Date', ascending=False)
# Display the data
print("US Indices Data:")
print(us_data)
print("\nUK Indices Data:")
print(uk_data)
```

```python
weekly_data = yf.download('^GSPC', interval='1wk', start='2020-01-01', end='2024-08-16')
weekly_data.sort_values(by='Date', ascending=False)
```
