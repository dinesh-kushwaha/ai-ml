```python
from nsetools import Nse
import requests

# Initialize NSE object
nse = Nse()

# Specify the desired expiration date
expiration_date = "29-Aug-2024"

# Fetch the option chain data
url = f"https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY&expiryDate={expiration_date}"
headers = {'User-Agent': 'Mozilla/5.0'}
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    option_chain_data = response.json()
    print(option_chain_data)
else:
    print("Failed to fetch data. Please check the expiration date or try again later.")
```
